```python
category_subscriber_mean = df.groupby('category')['subscribers'].mean()
category_subscriber_mean
```

```python
category_subscriber_mean = category_subscriber_mean.sort_values(ascending = False)
```

```python
category_subscriber_mean.head().plot(kind = 'bar')
plt.xlabel('Category')
plt.ylabel('Mean')
plt.title('Mean of Category subscriber ')
plt.show()
```

```python
channel_type_counts.tail().plot(kind = 'pie', autopct = '%1.2f%%')
plt.title('Channel Type Distribution')
plt.show()
```

```python
sns.boxplot(x = 'channel_type', y = 'highest_monthly_earnings', data = df)
plt.title('Distribution of Highest Monthly Earnings by channel types')
plt.xticks(rotation = 90)
plt.show()
```

```python
df.groupby(['category','Country'])['subscribers'].sum().unstack().plot(kind = 'bar', stacked = True)

plt.title('Subscribers Breakdown by category and country')
plt.xlabel('Category')
plt.ylabel('total subscribers')
plt.legend(title = 'Country', bbox_to_anchor = (1.05,1), loc = 'upper left')
```

```python
plt.scatter(df['Gross tertiary education enrollment (%)'],df['subscribers'], alpha = 0.5)
plt.title('Scatter plot of gross tertiary education enrollment vs subscribers')
plt.xlabel('Gross tertiary education enrollment')
plt.ylabel('susbcribers')
plt.show()
```

```python
df.groupby('channel_type')['Unemployment rate'].mean().sort_values(ascending = True).plot(kind = 'bar')

plt.title('Unemployment mean by various channel type')
plt.xlabel('channel type')
plt.ylabel('Umemployment rate')
```

```python
df.groupby(['created_year','channel_type'])['subscribers'].mean().unstack().tail().plot(kind = 'bar', width = 1)
plt.title('Avg subscriber by created year ')
plt.xlabel('Created Year')
plt.ylabel('Average subscribers')
plt.show()
```

```python
df.groupby(['created_year','created_month'])['subscribers'].sum().unstack().plot(kind = 'area')
plt.legend(title = 'Month', bbox_to_anchor = (1.05,1), loc = 'upper left')
```

```python
df.groupby(['created_year','Country'])['highest_yearly_earnings'].mean().unstack().head().plot(kind = 'line', marker = 'o')
plt.title('Highest Yearly earnings by created year & country')
plt.legend(title = 'Country', bbox_to_anchor = (1.05,1), loc = 'upper left')
```
